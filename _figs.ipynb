```python
df = data.get_data(use_processed=True, dset='orig') # all_data=True
df = df[df['valid'] == 1] # exclude test cells, short/long tracks, hotspots
outcome_def = 'y_consec_thresh'
```

```python
outcome = df[outcome_def]
R, C = 1, 3
plt.figure(figsize=(6, 2))
plt.subplot(R, C, 1)
plt.hist(df['lifetime'][outcome == 1], label='aux+', alpha=1, color=cb, bins=25)
plt.hist(df['lifetime'][outcome == 0], label='aux-', alpha=0.7, color=cr, bins=25)
plt.xlabel('Lifetime')
plt.ylabel('Count')

plt.subplot(R, C, 2)
plt.hist(df['X_max'][outcome == 1], label='aux+', alpha=1, color=cb, bins=25)
plt.hist(df['X_max'][outcome == 0], label='aux-', alpha=0.7, color=cr, bins=25)
plt.xlabel('Clathrin max amplitude')
plt.ylabel('Count')

plt.subplot(R, C, 3)
plt.hist(df['fall'][outcome == 1], label='aux+', alpha=1, color=cb, bins=25)
plt.hist(df['fall'][outcome == 0], label='aux-', alpha=0.7, color=cr, bins=25)
plt.xlabel('Fall')
plt.ylabel('Count')

plt.tight_layout()
plt.legend()
plt.savefig('figs/fig_hists.pdf')
plt.show()
```
